### Example 2 for the usage of Rela²x. 
### See README.md for more information.

- Two ¹H nuclei.

- Dipole-dipole coupling.

- Rank $l = 2$.

Import Rela²x.

```python
from rela2x import *
```

Define the spin system.

```python
spin_system = ['1H', '1H']
```

Define the incoherent interactions.

```python
dd_couplings = [[0, 1],
                [0, 0]]
intrs = {'DD': ('2', dd_couplings, [2])}
```

Compute the relaxation superoperator in the direct product basis of Cartesian spin operators and create a RelaxationSuperoperator object.

```python
R = R_object_in_prodop_basis(spin_system, intrs, basis='C')
```

Display $R$ and visualize.

```python
display(R.op)
R.visualize(basis_symbols=R.basis_symbols)
```

Check the auto-relaxation rate of $\hat S_{z}^{(1)}$, and the cross-relaxation rate between operators $\hat S_{z}^{(1)}$ and $\hat S_{z}^{(2)}$

```python
rate1 = R.rate('1z')
rate2 = R.rate('1z', '2z')

display(rate1)
display(rate2)
```

Use isotropic rotational diffusion model.

```python
R.to_isotropic_rotational_diffusion()
display(R.op)
```

Check again the auto-relaxation rate of $\hat S_{z}^{(1)}$, and the cross-relaxation rate between operators $\hat S_{z}^{(1)}$ and $\hat S_{z}^{(2)}$

```python
rate1 = R.rate('1z')
rate2 = R.rate('1z', '2z')

display(rate1)
display(rate2)
```

Compute the (Solomon) equations of motion and save in LaTeX format.

(Remove the comment if you want to save the file)

```python
eoms = equations_of_motion(R.op, R.basis_symbols, included_operators=[3, 12])
display(eoms)
# equations_of_motion_to_latex(eoms, 'dd_test')
```

Do the same calculation but now using the quantum mechanical (Lindbladian) formalism of relaxation.

```python
# Set quantum mechanical/Lindbladian relaxation theory
set_relaxation_theory('qm')

# Compute the relaxation superoperator in the direct product basis
R = R_object_in_prodop_basis(spin_system, intrs, basis='C')

# Display and visualize
display(R.op)
R.visualize(basis_symbols=R.basis_symbols)

# Compute the equations of motion
eoms = equations_of_motion(R.op, R.basis_symbols, included_operators=[3, 12])
display(eoms)
# equations_of_motion_to_latex(eoms, 'dd_test')
```
